```python
from pathlib import Path

sample_texts={"data/text_files/langchain_intro.txt": "LangChain is a popular open-source framework designed to make it easier to build applications with large language models (LLMs). Its modular design allows developers to chain together various components – such as prompt templates, memory, and external tools – enabling powerful, flexible AI-driven workflows."}

for filepath, content in sample_texts.items():
    path = Path(filepath)
    path.parent.mkdir(parents=True, exist_ok=True)
    with path.open('w', encoding='utf-8') as f:
        f.write(content)

print("Sample text files created successfully.")
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("data/text_files/langchain_intro.txt", encoding="utf-8")

documents = loader.load()

print(documents)
```

```python
from langchain_community.document_loaders import DirectoryLoader    

## Load all .txt files in the directory
dire_loader = DirectoryLoader("../data/text_files", glob="**/*.txt", show_progress=True)

documents = dire_loader.load()
print(documents)
```

```python
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader

## Load all .txt files in the directory
dire_loader = DirectoryLoader("../data/pdf", glob="**/*.pdf", loader_cls=PyMuPDFLoader,show_progress=True)

pdf_documents = dire_loader.load()
print(pdf_documents)
```

```python
from langchain_community.document_loaders import csv_loader

c_loader = DirectoryLoader("../data/csv", glob="**/*.csv"
                           , loader_class=csv_loader, show_progress=True)

csv_document = c_loader.load()
print(csv_document)
```
